## Define State

## Define Nodes

```python
from typing import TypedDict, Literal

class ItemState(TypedDict):
    user_request: str | None
    item_id_nbr: int | None
    item_loc_state: str | None
    item_loc_zip: str | None

    # Generated content
    draft_response: str | None

from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama
from typing import TypedDict

# ---- Create the Ollama model ----
llm = ChatOllama(
    model="nemotron-3-super:cloud",
    temperature=0.2,
)

# ---- Node function ----
def find_item(state: ItemState):
    result = llm.invoke(state["user_request"])
    return {'draft_response': result.content}

# ---- Build the graph ----
graph = StateGraph(ItemState)
graph.add_node("Find Item", find_item)
graph.set_entry_point("Find Item")
graph.add_edge("Find Item", END)

app = graph.compile()

out = app.invoke({'user_request': 'Say hello to the nice people'})

print(f'Response is {out["draft_response"]}')
```
